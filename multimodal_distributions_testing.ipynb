```python
import numpy as np
import matplotlib.pyplot as plt

import numpy.typing as npt
from typing import Callable
```

```python
def get_Z(xy: npt.Arraylike, f: Callable)-> nd.array:
    '''
    Take a grid of xy-tuples and calculates heights z
    xy: a numpy array (nd.array)
    f: a function defining the z values given x an y
    '''
    Nx = xy.shape[0]
    Ny = xy.shape[1]
    z = np.zeros((Nx,Ny))
    for i in range(Nx):
        for j in range(Ny):
            x,y =xy[i,j,0], xy[i,j,1]
            z[i,j] = f(x,y)
    return z
```

```python
N    = 200
X    = np.linspace(-6, 6, N)
Y    = np.linspace(-6, 6, N)
X, Y = np.meshgrid(X, Y)
pos  = np.dstack((X, Y))
s1 = lambda x,y: np.sin(2*np.pi*x*y/6) + np.cos(2*np.pi*y**2/6)
Z = get_Z(pos, s1)**2

Z = Z/np.sum(Z) # normalize over support xy

fig, ax = plt.subplots(figsize=(7,7))
cf = ax.contourf(X, Y, Z)
fig.colorbar(cf, ax=ax)
plt.show()
```
